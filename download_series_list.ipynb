```python
from nbiatoolkit import NBIAClient
from pathlib import Path
```

```python
dataset = 'CPTAC-CCRCC'
data_path = Path('/home/bhkuser/bhklab/radiomics/PublicDatasets')
src_data = data_path / 'srcdata'
proc_data = data_path / 'procdata'

download_dir = src_data / 'Abdomen' /  f'TCIA_{dataset}' / 'images' 

series_ids_file = src_data / 'Abdomen' / f'TCIA_{dataset}' / 'metadata/seg_series_ids.csv'
```

```python
with open(series_ids_file, 'r') as f:
    series_ids_list = [line.strip() for line in f]

series_ids_list = list(set(series_ids_list))  # Remove duplicates
```

```python
len(series_ids_list)
```

```python
with NBIAClient(return_type="dataframe") as client:
    series = client.getSeries(
        SeriesInstanceUID=series_ids_list[0]
    )
series
```

```python
file_pattern = '%Modality/%PatientID/%StudyInstanceUID/%SeriesInstanceUID/%InstanceNumber.dcm'
n_parallel = 8

with NBIAClient(return_type="dataframe") as client:
    client.downloadSeries(
        SeriesInstanceUID=series_ids_list,
        downloadDir=download_dir,
        filePattern=file_pattern,
        nParallel=n_parallel,
        Progressbar=True
    )
```
